# Explore single-cell data from Larsson et al. (2019)

mRNA content of 224 primary mouse fibroblasts was extracted using the Smart-seq2 protocol

https://www.nature.com/articles/s41586-018-0836-1

#### Load libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

#### Load data (as obtained preprocessed per allele from the paper)

```python
maternal_allele = pd.read_csv('../data/SS3_c57_UMIs_concat.csv', index_col=0)
paternal_allele = pd.read_csv('../data/SS3_cast_UMIs_concat.csv', index_col=0)
total_data = maternal_allele + paternal_allele
```

#### The object total_data contains mRNA counts for 10'727 genes across the 224 cells

```python
total_data
```

#### Visualize mRNA distribution for some key genes

For the Zyx gene

```python
gene = 'Zyx'
zyx_data = total_data.loc[gene][pd.notnull(total_data.loc[gene])]

plt.hist(zyx_data, bins=40, label='total', color='blue')
plt.xlabel("Total Expression")
plt.ylabel("Cells")
plt.tight_layout()
plt.show()
```
